```python
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
data = pd.read_csv('transport_data.csv')
local_route = data['Local Route']
light_rail = data['Light Rail']
peak_service = data['Peak Service']
rapid_route = data['Rapid Route']
school = data['School']
```

```python
def forecast_next_7_days(series):
    model = ARIMA(series, order=(5, 1, 0))  
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=7)
    return forecast
```

```python
local_route_forecast = forecast_next_7_days(local_route)
light_rail_forecast = forecast_next_7_days(light_rail)
peak_service_forecast = forecast_next_7_days(peak_service)
rapid_route_forecast = forecast_next_7_days(rapid_route)
school_forecast = forecast_next_7_days(school)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(local_route.index[-30:], local_route[-30:], label='Local Route', color='blue')
plt.plot(light_rail.index[-30:], light_rail[-30:], label='Light Rail', color='orange')
plt.plot(peak_service.index[-30:], peak_service[-30:], label='Peak Service', color='green')
plt.plot(rapid_route.index[-30:], rapid_route[-30:], label='Rapid Route', color='red')
plt.plot(school.index[-30:], school[-30:], label='School', color='purple')
plt.title('Transport Service Demand')
plt.xlabel('Date')
plt.ylabel('Demand')
plt.legend()
plt.show()
```
